# The below script will apply SMOTE(Synthetic Minority Oversampling Technique) to balance the dataset.

```python
import pandas as pd
from imblearn.over_sampling import SMOTE

# Load the dataset
file_path = 'anemia_dataset-Copy1.csv'  # Update with your actual file path
anemia_data = pd.read_csv(file_path)

# Drop unnecessary "Unnamed" columns if present
anemia_data = anemia_data.loc[:, ~anemia_data.columns.str.contains('^Unnamed')]

# Separate features and target
X = anemia_data[['%Red Pixel', '%Green pixel', '%Blue pixel', 'Hb']]
y = anemia_data['Anaemic'].apply(lambda x: 0 if x == "No" else 1)  # Encode "No" as 0 and "Yes" as 1

# Define the target count to balance the classes
desired_count = y.value_counts()[1]  # The count of "anemic" samples

# Apply SMOTE to make classes equal
smote = SMOTE(sampling_strategy="auto", k_neighbors=5, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Combine synthetic data with the original data for a final dataset
balanced_anemia_data = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=['Anaemic'])], axis=1)
balanced_anemia_data['Anaemic'] = balanced_anemia_data['Anaemic'].apply(lambda x: "No" if x == 0 else "Yes")

# Save to a new CSV file
balanced_anemia_data.to_csv('balanced_anemia_dataset.csv', index=False)

# Check the updated class distribution
print(balanced_anemia_data['Anaemic'].value_counts())
print("Balanced dataset shape:", balanced_anemia_data.shape)
```

```python
data=pd.read_csv("balanced_anemia_dataset.csv")
data
```

```python
data.drop(["Anaemic"],inplace=True,axis=1)

data
```

# dropping null values

```python
data.dropna(axis=1,inplace=True)
```

# scaling values from 0 to 1 from original values

```python
input_scale=MinMaxScaler()
output_scale=MinMaxScaler()
input_scale.fit(data.drop(["Hb"],axis=1))
output_scale.fit(np.array(data.Hb).reshape(-1,1))
```

```python
np.array(data.Hb).reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(data.drop(["Hb"],axis=1),data.Hb,test_size=0.15,random_state=44)
x_train,y_train
```

```python

x_train,x_test=input_scale.transform(x_train),input_scale.transform(x_test)
y_train,y_test=output_scale.transform(np.array(y_train).reshape(-1,1)),output_scale.transform(np.array(y_test).reshape(-1,1))
x_train,y_train
```

```python
dir(input_scale)
```

```python
model=keras.Sequential([
    layers.Dense(150,input_shape=(3,),activation="relu"),
    #layers.Dense(100,activation="relu"),
    #layers.Dense(50,activation="relu"),
    #layers.Dense(50,activation="relu"),
    layers.Dense(75,activation="relu"),
    layers.Dense(25,activation="relu"),
    layers.Dense(15,activation="relu"),
    layers.Dense(5,activation="relu"),
    layers.Dense(1,activation="linear")

]
)
```

```python
from keras.optimizers import Adam,SGD,RMSprop,Adagrad,Adamax
model.compile(optimizer=Adagrad(learning_rate=0.001),loss="mean_absolute_error",metrics=["mean_absolute_error"])
```

# configuring early stopping to get best weights fro  training

```python
from keras.callbacks import EarlyStopping
es=EarlyStopping(monitor='val_loss',patience=100,restore_best_weights=True)
print(model.summary())
```

```python
training_MAE=model.fit(x_train,y_train,epochs=2000,validation_split=0.2,verbose=1,callbacks=[es])
```

```python
training_MSE=model.fit(x_train,y_train,epochs=2000,validation_split=0.2,verbose=1,callbacks=[es])
```

# making predictions

```python
pred=output_scale.inverse_transform(model.predict(x_test))

pred1=model.predict(x_test)
pred,pred1
```

# evaluating metrics

```python
import numpy as np
rmse = np.sqrt(np.mean((np.array(y_test) -pred1) ** 2))
mse=np.mean((np.array(y_test) - pred1) ** 2)
mae = np.mean(np.abs(np.array(y_test) - pred1))
mse,rmse,mae
```
